```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def Lndp(series,index):
    mat=series
    code=[]
    m=8
    start=int(index-(m/2))
    end=int(index+(m/2))
    #print (start,end)
    j=0
    for i in range(start,end):
        x=mat[i]-mat[i+1]
        if (x>=0):
            code.append(1)
        else:
            code.append(0)
    #j=j+1 
    #r=j
    #j=0
    #print(code)
    #print(type(code))
    return code
```

```python
def code_to_binary(series,code):
    j=0
    x=0
    while (j<len(code)):
        x=x+code[j]*(2**j)
        j=j+1
    return x
```

```python
f=np.concatenate((a,d,e))
```

```python
y=np.zeros(shape=(300,1))
for i in range(100,200):
    y[i]=1
print(y.shape)
for i in range(200,300):
    y[i]=2
print(y[7],y[123],y[234])    
```

```python
from keras.utils import to_categorical
y_binary = to_categorical(y)
print(y_binary)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(f, y_binary, test_size=0.10)
print(X_train.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model= Sequential()
model.add(Dense(255,input_dim=255,activation='relu'))
model.add(Dense(127,activation='relu'))
model.add(Dense(127,activation='relu'))
model.add(Dense(3,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
#model.fit(X_train,y_train,epochs=15,batch_size=10)
```

```python
#scores=model.evaluate(X_test,y_test)
#print("\n%s:%.2f%%"%(model.metrics_names[1],scores[1]*100))
```

```python
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint
from keras.models import load_model
from matplotlib import pyplot
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1,patience=200)
mc = ModelCheckpoint('E:\\anilaka_project\\best_model_a_d_e.h5', monitor='val_accuracy', 
                     mode='max', verbose=1, save_best_only=True)
# fit model
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), 
                    epochs=100, verbose=0, callbacks=[es,mc])
saved_model = load_model('E:\\anilaka_project\\best_model_a_d_e.h5')
# evaluate the model
_, train_acc = model.evaluate(X_train, y_train, verbose=0)
_, test_acc = model.evaluate(X_test, y_test, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
# plot training history
pyplot.title("A-D-E ( LNDP )classifier")
pyplot.xlabel('epochs')
pyplot.ylabel('loss')
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
pyplot.show()
```
